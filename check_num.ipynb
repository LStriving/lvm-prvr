```python
import json
with open("video_info.json")as f:
    video_info = json.load(f)
```

```python
video_info = {i['video_name'][:-4]:i for i in video_info}
```

```python
from tqdm import tqdm
import os

actual_num = {}
videos = os.listdir("extracted336")
for video in tqdm(videos):
    actual_num[video] = len(os.listdir("extracted336/" + video))
```

```python
video_info  
```

```python
import math
for vid in actual_num:
    if actual_num[vid] != math.ceil(video_info[vid]["video_duration"] // 2 / 4):
        print(vid, actual_num[vid], math.ceil(video_info[vid]["video_duration"] // 2 / 4))
```
